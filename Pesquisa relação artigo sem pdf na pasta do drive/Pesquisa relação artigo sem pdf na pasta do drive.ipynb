```python
from google.colab import drive
import os

# Limpar tela do ambiente juppter
from IPython.display import clear_output

import time


# montar drive
drive.mount('/content/drive')

# caminho das pastas
# Drive  "/content/drive/Shareddrives/DTRI - BOLSISTAS/Analíticas/Base Dimensions/2022 - Rebeca Natalia/BAE_fazendo/BAE_pdfs"
# Saida lista de nomes no drive "/content/drive/MyDrive/Bc/Scrips/nomes.txt"
# Saida lista de relaçao de nomes sem arquivos "/content/drive/MyDrive/Bc/Scrips/nomes_sem_arquivos.txt"




################################################################################
#               Pesquisa relação artigo sem pdf na pasta do drive              #
################################################################################
# Função para colocar .pdf no final de cada nome dos arquivos
def aplica_pdf (lista_nome):
  for i in range(len(lista_nome)):
    lista_nome[i] = lista_nome[i] + ".pdf"

  return lista_nome

# Pega dados de input de usuário
def pega_dados():
  lista_nome_excel = input("Cole os valores da coluna Nome do arquivo: ").split(" ")

  # Limpar lista
  while True:
    aux = 0
    for nome_excel in lista_nome_excel:

      # Verifica se o elemneto é vazio se for joga para o final da lista e apaga ele
      if nome_excel == "":
        elemento = lista_nome_excel.pop(aux)
        lista_nome_excel.append(elemento)
        del lista_nome_excel[-1]
        break
      else: # Nome copia e cola excel não é vazio
        aux += 1
    if aux == len(lista_nome_excel):
      break

  lista_nome_excel = aplica_pdf(lista_nome_excel)

  return lista_nome_excel

def pesquisa_nomes(pasta_drive, arquivo_saida_nomes, arquivo_saida_relacao, copia_cola_excel):
  # Primeiro limpamos os arquivos de saidas
  limpa_lista(arquivo_saida_nomes)
  limpa_lista(arquivo_saida_relacao)

  # Segundo pegamos os dados do drive
  lista_nome_drive = puxa_nomes(pasta_drive, arquivo_saida_nomes)
  # Quarto rodamos um script de busca nome por nome

  # pega o primeiro nome da lista de nomes do copia e cola do excel e verifica com todos os nomes do drive
  # passa para o segundo nome, depois para o terceiro e assim até passar por todos os nomes do copia e cola do excel
  for nome_excel in copia_cola_excel:
    verificacao = True
    for nome_drive in lista_nome_drive:
      if nome_excel == nome_drive:
        verificacao = True
        break

      else:
        verificacao = False
    if verificacao == False:
      nome_excel_sem_pdf = nome_excel.replace(".pdf", "")
      with open(arquivo_saida_relacao, "a") as f:
        f.write(nome_excel_sem_pdf + "\n")

  print("####################################")
  print("#                                  #")
  print("#                                  #")
  print("#       Pesquisa concluida.        #")
  print("#                                  #")
  print("#                                  #")
  print("####################################")
  print("")
  print("")
  print("")
  print("Arquivo salvo em: ", arquivo_saida_relacao)

def menu_relacao_pdf():
  estado = 2

  pasta_drive = "/content/drive/Shareddrives/DTRI - BOLSISTAS/Analíticas/Base Dimensions/2022 - Rebeca Natalia/BAE_fazendo/BAE_pdfs"
  arquivo_saida_nomes = "/content/drive/MyDrive/Bc/Scrips/nomes.txt"
  arquivo_saida_relacao = "/content/drive/MyDrive/Bc/Scrips/nomes_sem_arquivos.txt"
  copia_cola_excel = ""
  """
  pasta_drive = ""
  arquivo_saida_nomes = ""
  arquivo_saida_relacao = ""
  copia_cola_excel = ""
  """
  while estado != 0:
    limpar_tela()
    print("####################################")
    print("")
    print("")
    print("1, Adicionar pasta para pesquisa.")
    print("2, Acicionar arquivo de saida da lista de nomes.")
    print("3, Adicionar arquivo de saida da relação de nomes sem arquivos.")
    print("4, Adicionar nomes do copia e cola do excel.")
    print("0, sair.")
    print("")
    print("Pasta do drive para pesquisa: ", pasta_drive)
    print("Arquivo para salvar nomes.txt", arquivo_saida_nomes)
    print("Arquivo para salvar relação de nomes sem arquivos", arquivo_saida_relacao)
    if copia_cola_excel != "":
      print("Lista de nomes do excel copiado.")
    else:
      print("Lista de nomes do excel não copiado.")
    print("")
    print("")
    print("")
    print("")
    print("5, Pesquisar")

    try:
      estado = int(input("Opção escolhida: "))

      if estado == 1:
        print("####################################")
        print("")
        print("")
        pasta_drive = montegem_caminho()

      elif estado == 2:
        print("####################################")
        print("")
        print("")
        arquivo_saida_nomes = montegem_caminho()

      elif estado == 3:
        print("####################################")
        print("")
        print("")
        arquivo_saida_relacao = montegem_caminho()

      elif estado == 4:
        print("####################################")
        print("")
        print("")
        copia_cola_excel = pega_dados()

      elif estado == 5:
        if pasta_drive != "" and arquivo_saida_nomes != "" and arquivo_saida_relacao != "" and copia_cola_excel != "":
          pesquisa_nomes(pasta_drive, arquivo_saida_nomes, arquivo_saida_relacao, copia_cola_excel)
        else:
          print("####################################")
          print("#                                  #")
          print("#    Preencha todos os campos.     #")
          print("#                                  #")
          print("####################################")
          print("")
          print("")
          print("")

      elif estado == 6:
          print("################################")
          print("#     #")
          print("################################")
          print("")
          print("")
          print("")

      elif estado == 0:
        break
      else:
        print("Opção invalida.")

    except KeyboardInterrupt:
        print("\nEncerrando...")
        break

################################################################################
#                    Números de arquivos em uma pasta drive                    #
################################################################################

def menu_n_arquivos():
  estado = 9
  pasta_montada = ""
  while estado != 0:
    limpar_tela()
    print("####################################")
    print("")
    print("Precione:")
    print("0, sair.")
    print("1, adicionar pasta.")
    print("")
    print("")
    print("")
    print("5, Pesquisar números de arquivos.")
    print("")
    print("")
    print("")
    print("Pasta montada: ", pasta_montada)

    try:

      estado = int(input("Opção escolhida: "))

      if estado == 0:
        break

      if estado == 1:
        pasta_drive = montegem_caminho()

      if estado == 5:
        print("####################################")
        print("")
        print("")
        # Lista dos nomes dos arquivos
        lista_arquivos = os.listdir(pasta_drive)

        # Numeros de arquivos no pasta do drive
        numero_arquivos = len(lista_arquivos)

        # Devolvendo o numeros de nomes
        print(f"O numero de arquivos é: {numero_arquivos}")

      else:
        print("Opção invalida.")

    except KeyboardInterrupt:
        print("\nEncerrando...")
        break
################################################################################
#                             Funções Gerais                                   #
################################################################################

# Monta o caminho para um arquivo ou pasta
def montegem_caminho():
  montagem_caminho = input("Digite o caminho da pasta/arquivo: ")
  return montagem_caminho


# Limpa arquivo de saida para salvar
def limpa_lista(arquivo_saida):
  with open(arquivo_saida, "w") as f:
    f.write("")

# Pega os nomes da pasta dada
def puxa_nomes(pasta_drive, arquivo_saida):
  # Lista dos nomes dos arquivos
  lista_nome = os.listdir(pasta_drive)

  # Salvar os nomes dos arquivos em um arquivo .txt separados por "\n"
  with open(arquivo_saida, "w") as f:
    for file in lista_nome:
            f.write(file + "\n")  # Adiciona ";" e quebra de linha após cada nome
  return lista_nome

# Limpa a tela para printar o proximo menu
def limpar_tela():
    time.sleep(1.5)
    clear_output(wait=True)
    time.sleep(1.5)

################################################################################
#                            Função Principal                                  #
################################################################################

def main ():

  # Fica em loop até a opção for 0
  estado = 9
  while estado != 0:
    # Imprime duas opções, 0 sair ou 1 continuar
    limpar_tela()
    print("####################################")
    print("")
    print("")
    print("0, sair.")
    print("1, Pesquisa relação artigo sem pdf na pasta do drive.")
    print("2, Números de arquivos em uma pasta drive.")
    print("")
    print("")
    print("")

    try:

      estado = int(input("Digite: "))
      if estado == 1:
        menu_relacao_pdf()

      elif estado == 2:
        menu_n_arquivos()

      elif estado == 0:
        break
      else:
        print("Opção invalida.")

    except KeyboardInterrupt:
        print("\nEncerrando...")
        break

main()
```
